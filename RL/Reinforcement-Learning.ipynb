```python
# RL Assignment Code Structure: Full Implementation - Blackjack SARSA, CartPole VI, PI, SARSA, DQN

import numpy as np
import gymnasium as gym
import random
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

"""
Section 1: Blackjack SARSA
"""
def run_blackjack_sarsa(num_episodes=10000, alpha=0.1, gamma=1.0, epsilon=1.0, epsilon_decay=0.9995, epsilon_min=0.1):
    env = gym.make("Blackjack-v1")
    Q = {}  # state-action value dictionary

    def get_Q(state, action):
        return Q.get((state, action), 0.0)

    rewards = []
    for episode in range(num_episodes):
        state, _ = env.reset()
        action = env.action_space.sample() if random.random() < epsilon else np.argmax([get_Q(state, a) for a in range(env.action_space.n)])
        total_reward = 0
        done = False

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            next_action = env.action_space.sample() if random.random() < epsilon else np.argmax([get_Q(next_state, a) for a in range(env.action_space.n)])

            td_target = reward + gamma * get_Q(next_state, next_action)
            td_delta = td_target - get_Q(state, action)
            Q[(state, action)] = get_Q(state, action) + alpha * td_delta

            state, action = next_state, next_action
            total_reward += reward

        rewards.append(total_reward)
        epsilon = max(epsilon_min, epsilon * epsilon_decay)
    return Q, rewards


"""
Section 2: CartPole Discretization Helpers
"""
def discretize_state(state, bins):
    upper_bounds = [4.8, 5, 0.418, 5]
    lower_bounds = [-4.8, -5, -0.418, -5]
    ratios = [(state[i] + abs(lower_bounds[i])) / (upper_bounds[i] - lower_bounds[i]) for i in range(len(state))]
    new_state = [int(round((bins[i] - 1) * ratios[i])) for i in range(len(state))]
    new_state = [min(bins[i] - 1, max(0, new_state[i])) for i in range(len(state))]
    return tuple(new_state)

"""
Section 3: CartPole SARSA (Tabular)
"""
def run_cartpole_sarsa(num_episodes=10000, bins=[10, 10, 10, 10], alpha=0.1, gamma=0.99, epsilon=1.0, epsilon_decay=0.999, epsilon_min=0.01):
    env = gym.make("CartPole-v1")
    Q = {}

    def get_Q(state, action):
        return Q.get((state, action), 0.0)

    episode_lengths = []
    for episode in range(num_episodes):
        state, _ = env.reset()
        state = discretize_state(state, bins)
        action = env.action_space.sample() if random.random() < epsilon else np.argmax([get_Q(state, a) for a in range(env.action_space.n)])
        done = False
        t = 0

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            next_state = discretize_state(next_state, bins)
            next_action = env.action_space.sample() if random.random() < epsilon else np.argmax([get_Q(next_state, a) for a in range(env.action_space.n)])
            td_target = reward + gamma * get_Q(next_state, next_action)
            td_delta = td_target - get_Q(state, action)
            Q[(state, action)] = get_Q(state, action) + alpha * td_delta
            state, action = next_state, next_action
            t += 1
        episode_lengths.append(t)
        epsilon = max(epsilon_min, epsilon * epsilon_decay)
    return Q, episode_lengths

"""
Section 4: Deep Q Network (CartPole)
"""
class DQN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(DQN, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, output_dim)
        )

    def forward(self, x):
        return self.fc(x)

"""
Section 5: DQN CartPole Trainer
"""
def run_cartpole_dqn(num_episodes=500, gamma=0.99, epsilon_start=1.0, epsilon_end=0.01, epsilon_decay=500, buffer_size=50000, batch_size=64):
    env = gym.make("CartPole-v1")
    n_actions = env.action_space.n
    state_dim = env.observation_space.shape[0]
    policy_net = DQN(state_dim, n_actions)
    target_net = DQN(state_dim, n_actions)
    target_net.load_state_dict(policy_net.state_dict())
    target_net.eval()
    optimizer = optim.Adam(policy_net.parameters(), lr=1e-3)
    memory = []
    rewards = []
    steps_done = 0

    def select_action(state, epsilon):
        if random.random() < epsilon:
            return env.action_space.sample()
        else:
            with torch.no_grad():
                return policy_net(torch.tensor(state, dtype=torch.float32)).argmax().item()

    for episode in range(num_episodes):
        state, _ = env.reset()
        total_reward = 0
        done = False
        while not done:
            epsilon = epsilon_end + (epsilon_start - epsilon_end) * np.exp(-1. * steps_done / epsilon_decay)
            action = select_action(state, epsilon)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            memory.append((state, action, reward, next_state, done))
            if len(memory) > buffer_size:
                memory.pop(0)

            if len(memory) >= batch_size:
                batch = random.sample(memory, batch_size)
                states, actions, rewards_b, next_states, dones = zip(*batch)
                states = torch.tensor(states, dtype=torch.float32)
                actions = torch.tensor(actions)
                rewards_b = torch.tensor(rewards_b, dtype=torch.float32)
                next_states = torch.tensor(next_states, dtype=torch.float32)
                dones = torch.tensor(dones, dtype=torch.bool)

                q_values = policy_net(states).gather(1, actions.unsqueeze(1)).squeeze()
                with torch.no_grad():
                    next_q_values = target_net(next_states).max(1)[0]
                    target = rewards_b + gamma * next_q_values * (~dones)

                loss = nn.functional.mse_loss(q_values, target)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            if steps_done % 100 == 0:
                target_net.load_state_dict(policy_net.state_dict())
            state = next_state
            total_reward += reward
            steps_done += 1
        rewards.append(total_reward)
    return rewards


"""
Section 6: CartPole Value Iteration
"""
def run_cartpole_value_iteration(gamma=0.99, theta=1e-6, bins=[10,10,10,10]):
    env = gym.make("CartPole-v1")
    n_actions = env.action_space.n
    Q = {}
    V = {}

    def get_value(state):
        return V.get(state, 0.0)

    states = [(i, j, k, l) for i in range(bins[0]) for j in range(bins[1]) for k in range(bins[2]) for l in range(bins[3])]
    for state in states:
        V[state] = 0.0
        for action in range(n_actions):
            Q[(state, action)] = 0.0

    while True:
        delta = 0
        for state in states:
            v = get_value(state)
            q_values = []
            for action in range(n_actions):
                env.reset()
                env.state = [0, 0, 0, 0]  # starting from neutral state for simplicity
                next_state, reward, done, _, _ = env.step(action)
                next_state = discretize_state(next_state, bins)
                q = reward + gamma * get_value(next_state)
                Q[(state, action)] = q
                q_values.append(q)
            V[state] = max(q_values)
            delta = max(delta, abs(v - V[state]))
        if delta < theta:
            break
    policy = {state: np.argmax([Q[(state, a)] for a in range(n_actions)]) for state in states}
    return policy, V

"""
Section 7: CartPole Policy Iteration
"""
def run_cartpole_policy_iteration(gamma=0.99, theta=1e-6, bins=[10,10,10,10]):
    env = gym.make("CartPole-v1")
    n_actions = env.action_space.n
    V = {}
    policy = {}

    states = [(i, j, k, l) for i in range(bins[0]) for j in range(bins[1]) for k in range(bins[2]) for l in range(bins[3])]
    for state in states:
        V[state] = 0.0
        policy[state] = env.action_space.sample()

    is_policy_stable = False
    while not is_policy_stable:
        # Policy Evaluation
        while True:
            delta = 0
            for state in states:
                v = V[state]
                action = policy[state]
                env.reset()
                env.state = [0, 0, 0, 0]
                next_state, reward, done, _, _ = env.step(action)
                next_state = discretize_state(next_state, bins)
                V[state] = reward + gamma * V[next_state]
                delta = max(delta, abs(v - V[state]))
            if delta < theta:
                break
        # Policy Improvement
        is_policy_stable = True
        for state in states:
            old_action = policy[state]
            q_values = []
            for action in range(n_actions):
                env.reset()
                env.state = [0, 0, 0, 0]
                next_state, reward, done, _, _ = env.step(action)
                next_state = discretize_state(next_state, bins)
                q = reward + gamma * V[next_state]
                q_values.append(q)
            policy[state] = np.argmax(q_values)
            if old_action != policy[state]:
                is_policy_stable = False
    return policy, V

"""
Section 8: FrozenLake Value Iteration
"""
def run_frozenlake_value_iteration(gamma=0.99, theta=1e-6):
    env = gym.make("FrozenLake-v1", is_slippery=True).unwrapped
    n_states = env.observation_space.n
    n_actions = env.action_space.n
    V = np.zeros(n_states)

    def one_step_lookahead(state, V):
        A = np.zeros(n_actions)
        for a in range(n_actions):
            for prob, next_state, reward, done in env.P[state][a]:
                A[a] += prob * (reward + gamma * V[next_state])
        return A

    iteration_count = 0
    while True:
        delta = 0
        iteration_count += 1
        for s in range(n_states):
            A = one_step_lookahead(s, V)
            best_action_value = np.max(A)
            delta = max(delta, np.abs(best_action_value - V[s]))
            V[s] = best_action_value
        if delta < theta:
            break

    # Extract policy after convergence
    policy = np.zeros(n_states, dtype=int)
    for s in range(n_states):
        A = one_step_lookahead(s, V)
        policy[s] = np.argmax(A)

    return policy, V, iteration_count


    ...


"""
Section 9: FrozenLake Policy Iteration
"""
def run_frozenlake_policy_iteration(gamma=0.99, theta=1e-6):
    env = gym.make("FrozenLake-v1", is_slippery=True).unwrapped
    n_states = env.observation_space.n
    n_actions = env.action_space.n
    policy = np.zeros(n_states, dtype=int)
    V = np.zeros(n_states)

    def policy_eval(policy, V):
        while True:
            delta = 0
            for s in range(n_states):
                v = 0
                a = policy[s]
                for prob, next_state, reward, done in env.P[s][a]:
                    v += prob * (reward + gamma * V[next_state])
                delta = max(delta, np.abs(v - V[s]))
                V[s] = v
            if delta < theta:
                break
        return V

    iteration_count = 0
    while True:
        iteration_count += 1
        V = policy_eval(policy, V)
        stable = True
        for s in range(n_states):
            old_action = policy[s]
            A = np.zeros(n_actions)
            for a in range(n_actions):
                for prob, next_state, reward, done in env.P[s][a]:
                    A[a] += prob * (reward + gamma * V[next_state])
            best_action = np.argmax(A)
            if best_action != old_action:
                stable = False
            policy[s] = best_action
        if stable:
            break

    return policy, V, iteration_count


```

```python
import pandas as pd

def moving_average(data, window=100):
    return pd.Series(data).rolling(window).mean()
```

```python
# BLACKJACK SARSA
Q_blackjack, rewards_blackjack = run_blackjack_sarsa(num_episodes=10000)
plt.plot(moving_average(rewards_blackjack, window=100))
plt.title("Blackjack SARSA Smoothed Rewards (100-episode average)")
plt.xlabel("Episode")
plt.ylabel("Average Reward")
plt.grid()
plt.show()
```

```python
# CARTPOLE SARSA (TABULAR)
Q_cartpole, episode_lengths = run_cartpole_sarsa(num_episodes=10000)
plt.plot(moving_average(episode_lengths, window=200))
plt.title("CartPole SARSA Smoothed Episode Lengths")
plt.xlabel("Episode")
plt.ylabel("Average Episode Length (200 ep avg)")
plt.ylim(0, 500)
plt.grid()
plt.show()
```

```python
# CARTPOLE DQN

dqn_rewards = run_cartpole_dqn(num_episodes=500)
plt.plot(moving_average(dqn_rewards, window=20))
plt.title("CartPole DQN Smoothed Episode Rewards")
plt.xlabel("Episode")
plt.ylabel("Average Reward (20 ep avg)")
plt.ylim(0, 500)
plt.grid()
plt.show()
```

```python
# CARTPOLE VALUE ITERATION
policy_vi, V_vi = run_cartpole_value_iteration()
plt.plot(sorted(V_vi.values()))
plt.title("CartPole Value Iteration State Values (Sorted)")
plt.xlabel("State Index")
plt.ylabel("Value")
plt.show()
```

```python
# CARTPOLE POLICY ITERATION
policy_pi, V_pi = run_cartpole_policy_iteration()
plt.plot(sorted(V_pi.values()))
plt.title("CartPole Policy Iteration State Values (Sorted)")
plt.xlabel("State Index")
plt.ylabel("Value")
plt.show()
```

```python
# FROZENLAKE VALUE ITERATION
policy_vi, V_vi, vi_iterations = run_frozenlake_value_iteration()
plt.bar(range(len(V_vi)), V_vi)
plt.title("FrozenLake Value Iteration: State Values")
plt.xlabel("State")
plt.ylabel("Value")
plt.grid()
plt.show()
print(f"Value Iteration converged in {vi_iterations} iterations")
```

```python
# FROZENLAKE POLICY ITERATION
policy_pi, V_pi, pi_iterations = run_frozenlake_policy_iteration()
plt.bar(range(len(V_pi)), V_pi)
plt.title("FrozenLake Policy Iteration: State Values")
plt.xlabel("State")
plt.ylabel("Value")
plt.grid()
plt.show()
print(f"Policy Iteration converged in {pi_iterations} iterations")
```
